# **Introduction**
In this report, I will analyze the financial data of Amazon (AMZN) using Python, with a focus on implementing a simple trading strategy based on two Simple Moving Averages (SMA). The goal is to evaluate the effectiveness of this strategy over the past two years. The strategy will be backtested using historical stock price data, and key metrics will be calculated to assess its performance.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
```

# **Data Collection**

For this analysis, I have used Yahoo Finance to download the historical stock price data of Amazon (AMZN) for the last two years, from January 1, 2022, to January 1, 2024. The dataset includes daily open, high, low, close prices, and volume, which are crucial for calculating the Simple Moving Averages (SMA) and backtesting the trading strategy.


```python

stock_data = yf.download('AMZN', start='2022-01-01', end='2024-01-01', interval='1d')
stock_data.head()
```

# **Strategy Description**

The strategy I implemented is based on two Simple Moving Averages (SMA):
- **Short-term SMA**: 20 days
- **Long-term SMA**: 50 days

The trading rules are defined as follows:
- **Buy Signal**: When the 20-day SMA crosses above the 50-day SMA.
- **Sell Signal**: When the 20-day SMA crosses below the 50-day SMA.

This crossover strategy is a popular method used in technical analysis to identify potential buy and sell points. The rationale behind this strategy is that the short-term trend (20-day SMA) crossing above the long-term trend (50-day SMA) indicates a bullish signal, suggesting it's a good time to buy. Conversely, when the short-term trend crosses below the long-term trend, it indicates a bearish signal, suggesting it's time to sell.


```python

stock_data['SMA_20'] = stock_data['Close'].rolling(window=20).mean()


stock_data['SMA_50'] = stock_data['Close'].rolling(window=50).mean()

stock_data.tail()
```

```python
# signal is 0 (no position), 1 (buy signal), -1 (sell signal)
stock_data['Signal'] = 0
stock_data['Signal'][20:] = np.where(stock_data['SMA_20'][20:] > stock_data['SMA_50'][20:], 1, 0)
stock_data['Signal'][20:] = np.where(stock_data['SMA_20'][20:] < stock_data['SMA_50'][20:], -1, stock_data['Signal'][20:])

stock_data['Daily_Return'] = stock_data['Close'].pct_change()
stock_data['Strategy_Return'] = stock_data['Signal'].shift(1) * stock_data['Daily_Return']

stock_data.tail()
```

# **Backtesting Results**

To evaluate the performance of the SMA crossover strategy, I backtested it on the historical data of Amazon with an initial capital of 1,00,000. The key metrics obtained from the backtest are as follows:

- **Total Returns**: -15112.16
- **Number of Trades**:19.0
- **Winning Trades**: 1147.37%
- **Losing Trades**: 1215.79%
- **Maximum Drawdown**:41831.12

The total returns generated by the strategy indicate the overall profit or loss over the two-year period. The number of trades reflects the frequency of buy/sell signals. The percentage of winning and losing trades provides insight into the accuracy of the signals. The maximum drawdown is an important risk metric that shows the largest peak-to-trough decline in the equity curve.


```python
initial_capital = 100000
stock_data['Cumulative_Strategy_Return'] = (1 + stock_data['Strategy_Return']).cumprod()

# equity curve (value of the portfolio)
stock_data['Equity_Curve'] = initial_capital * stock_data['Cumulative_Strategy_Return']

total_returns = stock_data['Equity_Curve'].iloc[-1] - initial_capital
number_of_trades = stock_data['Signal'].diff().abs().sum()
winning_trades = len(stock_data[stock_data['Strategy_Return'] > 0])
losing_trades = len(stock_data[stock_data['Strategy_Return'] < 0])
max_drawdown = stock_data['Equity_Curve'].max() - stock_data['Equity_Curve'].min()

print(f"Total Returns: {total_returns:.2f}")
print(f"Number of Trades: {number_of_trades}")
print(f"Winning Trades: {winning_trades / number_of_trades * 100:.2f}%")
print(f"Losing Trades: {losing_trades / number_of_trades * 100:.2f}%")
print(f"Maximum Drawdown: {max_drawdown:.2f}")
```

# **Visualization**

### 5.1 Stock Price with SMAs
- The first chart shows the daily closing price of Amazon along with the 20-day and 50-day SMAs. The crossover points highlight potential buy and sell signals.

### 5.2 Equity Curve
- The second chart represents the equity curve, showing how the value of the portfolio evolved over time based on the SMA strategy.


```python
# Plot stock price with SMAs
plt.figure(figsize=(14, 7))
plt.plot(stock_data['Close'], label='Close Price')
plt.plot(stock_data['SMA_20'], label='20-Day SMA')
plt.plot(stock_data['SMA_50'], label='50-Day SMA')
plt.title('Stock Price with SMAs')
plt.legend()
plt.show()

# Plot the equity curve
plt.figure(figsize=(14, 7))
plt.plot(stock_data['Equity_Curve'], label='Equity Curve')
plt.title('Equity Curve')
plt.legend()
plt.show()
```
